# WORLD ECONOMIC DATA ANALYSIS

This is an overview of World economic data that comprises of Corruption, Cost of living, richest Countries, Unemployment and Tourism datasets. A look through key insights within the datasets to find relationships and trends. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# CORRUPTION 

```python
df = pd.read_csv("/kaggle/input/world-economic-data/corruption.csv")
df
```

```python
#Checking shape of the data 
df.shape
```

```python
df.head()
```

```python
df.isnull
```

```python
df.info()
```

```python
corr = df.sort_values(by='corruption_index', ascending=False)
```

```python
corr.plot(x = 'country', y = 'corruption_index', kind='bar', title = "Corruption Index Per Country", rot = 90, figsize=(20, 10))
```

```python
annual = df.sort_values(by='annual_income', ascending=False)
```

```python
annual.plot(x = 'country', y = 'annual_income', kind='bar',title = "Annual Income per Country", rot = 90, figsize=(20, 10))
```

```python
df.plot(x = 'corruption_index', y = 'annual_income', kind='scatter',title ="Correlation between Corruption Index and Annual Income", rot = 90, figsize=(20, 10))
```

we see a negative correlation. There is a high annual income where there is less corruption.

# COST OF LIVING

```python
cost_of_living = pd.read_csv('/kaggle/input/world-economic-data/cost_of_living.csv')
cost_of_living
```

```python
cost_of_living.head()
```

```python
cost_of_living.shape
```

```python
cost_of_living.describe
```

```python
MI_cost_of_living = cost_of_living.sort_values(by = 'monthly_income', ascending = False)
MI_cost_of_living
```

```python
MI_cost_of_living.plot(x = 'country', y = 'monthly_income',title ="Cost of Living per Country", kind='bar', rot = 90, figsize=(20, 10))
```

```python
cost_of_living.plot(x = 'country', y = 'cost_index', title = "Cost Index per Country", kind='bar', rot = 90, figsize=(20, 15))
```

```python
PPI_cost_of_living = cost_of_living.sort_values(by='purchasing_power_index', ascending=False)
PPI_cost_of_living
```

```python
PPI_cost_of_living.plot(x = 'country', y = 'purchasing_power_index', kind='bar', rot = 90, figsize=(20, 15))
```

```python
cost_of_living.plot(x = 'monthly_income', y = 'cost_index', title = "Relation between Monthly Income and Cost Index", kind='scatter', rot = 90, figsize=(20, 15))
```

**The higher the salary, the higher the cost index.**

```python
sns.heatmap(cost_of_living.corr())
```

**The heat map shows how monthly income and the cost index interact, as well as monthly income and the purchasing power index.**
